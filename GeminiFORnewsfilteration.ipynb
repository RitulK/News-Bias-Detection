```python
from google.colab import userdata
from google import genai

GOOGLE_API_KEY = userdata.get("GOOGLE_API_KEY_2")
client = genai.Client(api_key=GOOGLE_API_KEY)
```

```python
from IPython.display import Markdown

MODEL_ID = "gemini-2.0-flash" # @param ["gemini-2.0-flash-lite","gemini-2.0-flash","gemini-2.5-pro-exp-03-25"] {"allow-input":true, isTemplate: true}

# Generate sample news articles covering different topics
response = client.models.generate_content(
    model=MODEL_ID,
    contents="Generate 3 different short news articles (separated by '---'): "
             "1) About politics, 2) About environment, 3) About economics, "
             "and 4) One unrelated article about sports",
)
articles = response.text.split('---')
Markdown("## Sample Articles:\n" + "\n\n---\n\n".join(articles))
```

```python
import enum
from typing_extensions import TypedDict  # in python 3.12 replace typing_extensions with typing

from google.genai import types


class Relevance(enum.Enum):
  WEAK = "weak"
  STRONG = "strong"

class Topic(TypedDict):
  topic: str
  relevance: Relevance


sys_int = """
Generate topics from text. Ensure that topics are general e.g. "Health".
Strong relevance is obtained when the topic is a core tenent of the content
and weak relevance reflects one or two mentions.
"""

response = client.models.generate_content(
    model=MODEL_ID,
    contents=articles,
    config=types.GenerateContentConfig(
        system_instruction=sys_int,
        response_mime_type="application/json",
        response_schema=list[Topic],
    )
)
```

```python
from IPython.display import display, HTML

# Simple classification function
def is_relevant(article):
    response = client.models.generate_content(
        model=MODEL_ID,
        contents=f"Does this article primarily discuss Politics, Environment, or Economics? "
                 f"Answer only 'YES' or 'NO'.\n\nArticle:\n{article}"
    )
    return response.text.strip().upper() == "YES"

# Process and display articles
display(HTML("<h2>Relevant Articles Highlighted:</h2>"))
for i, article in enumerate(articles, 1):
    if is_relevant(article):
        display(HTML(f"""
        <div style='border: 2px solid green; padding: 10px; margin: 10px;'>
            <h3>Article {i} (RELEVANT)</h3>
            <p>{article}</p>
        </div>
        """))
    else:
        display(HTML(f"""
        <div style='padding: 5px; margin: 5px; color: gray;'>
            <p>Article {i} (other topic)</p>
        </div>
        """))
```

```python
import json
from time import sleep

# Configuration
BATCH_SIZE = 4  # Number of articles to process at once
DELAY_SECONDS = 1  # Delay between batches to avoid rate limits

def process_batch(batch):
    """Process a batch of articles through Gemini API"""
    contents = [art['content'] for art in batch]

    response = client.models.generate_content(
        model=MODEL_ID,
        contents=f"""For each article below, classify into ONLY these categories:
        - Politics (government, elections, policies)
        - Environment (climate, nature, conservation)
        - Economics (markets, finance, trade)
        - Other

        Return JSON format: [{{"id": index, "category": category}}]

        Articles:\n{contents}"""
    )

    try:
        return json.loads(response.text)
    except json.JSONDecodeError:
        print("Failed to parse API response, retrying with individual requests")
        return None

def classify_single(article):
    """Fallback for single article classification if batch fails"""
    category = classify_article(article['content'])
    return {"id": 0, "category": category}  # Using id=0 since we're processing single article

# Load your JSON data
with open('filtered_news_articles.json') as f:
    news_data = json.load(f)

filtered_articles = []
failed_articles = []

# Process in batches
for i in range(0, len(news_data), BATCH_SIZE):
    batch = news_data[i:i+BATCH_SIZE]
    print(f"Processing batch {i//BATCH_SIZE + 1}/{(len(news_data)-1)//BATCH_SIZE + 1}")

    classifications = process_batch(batch)

    if classifications is None:
        # Fallback to individual processing if batch fails
        for article in batch:
            try:
                result = classify_single(article)
                if result['category'] in ['Politics', 'Environment', 'Economics']:
                    article['news_type'] = result['category']
                    filtered_articles.append(article)
            except Exception as e:
                print(f"Failed to process article {i}: {str(e)}")
                failed_articles.append(article)
    else:
        # Process batch results
        for j, classification in enumerate(classifications):
            if classification['category'] in ['Politics', 'Environment', 'Economics']:
                article = batch[j]
                article['news_type'] = classification['category']
                filtered_articles.append(article)

    sleep(DELAY_SECONDS)  # Be nice to the API

# Save results
with open('filtered_news.json', 'w') as f:
    json.dump(filtered_articles, f, indent=2)

print(f"\nProcessing complete!")
print(f"Successfully filtered {len(filtered_articles)} relevant articles")
print(f"Failed to process {len(failed_articles)} articles")

if failed_articles:
    with open('failed_articles.json', 'w') as f:
        json.dump(failed_articles, f, indent=2)
```

```python
import json
import time
from datetime import datetime
import os
import re

# Configuration
MAX_RETRIES = 2  # Reduced retries to avoid quota exhaustion
DELAY_BETWEEN_REQUESTS = 10  # Increased delay between requests
CONTENT_LIMIT = 2000  # Further reduce content length
SAVE_INTERVAL = 5  # Save progress every N articles

# Keywords for classification (fallback method)
TOPIC_KEYWORDS = {
    'Politics': ['election', 'government', 'minister', 'policy', 'law', 'political'],
    'Environment': ['climate', 'environment', 'pollution', 'conservation', 'wildlife', 'sustainability'],
    'Economics': ['economy', 'market', 'finance', 'trade', 'gdp', 'inflation']
}

def keyword_classification(text):
    """Fallback classification using keyword matching"""
    text_lower = text.lower()
    for topic, keywords in TOPIC_KEYWORDS.items():
        if any(keyword in text_lower for keyword in keywords):
            return topic
    return None

def classify_article(article, attempt=1):
    """Robust classification with multiple fallback methods"""
    try:
        # Pre-process content
        content = article['content'][:CONTENT_LIMIT]
        title = article['title'][:200]

        # Try API classification first
        response = client.models.generate_content(
            model=MODEL_ID,
            contents=f"""Classify this news article into ONLY these categories:
            - Politics (government, elections, policies)
            - Environment (climate, nature, conservation)
            - Economics (markets, finance, trade)
            - Other

            Return ONLY the category name. No explanations.

            Title: {title}
            Content: {content[:1000]}...""",
            generation_config={
                "max_output_tokens": 20,
                "temperature": 0.3
            }
        )

        # Clean and validate response
        category = response.text.strip()
        if category in ['Politics', 'Environment', 'Economics']:
            return category

        # If API returned 'Other' or invalid, try keyword matching
        return keyword_classification(content) or 'Other'

    except Exception as e:
        if attempt < MAX_RETRIES:
            wait_time = DELAY_BETWEEN_REQUESTS * attempt
            print(f"Attempt {attempt} failed. Waiting {wait_time}s...")
            time.sleep(wait_time)
            return classify_article(article, attempt + 1)

        # Final fallback to keyword matching
        print("API failed, using keyword fallback")
        return keyword_classification(article['content'])

# Initialize or load progress
if os.path.exists('progress.json'):
    with open('progress.json') as f:
        progress = json.load(f)
    print(f"Resuming from article {progress['last_processed'] + 1}")
else:
    progress = {
        'last_processed': -1,
        'filtered': [],
        'failed': []
    }

# Load news data
with open('filtered_news_articles.json') as f:
    news_data = json.load(f)

# Process articles
total_articles = len(news_data)
start_idx = progress['last_processed'] + 1

for i in range(start_idx, total_articles):
    article = news_data[i]
    print(f"\nProcessing article {i + 1}/{total_articles}: {article['title'][:50]}...")

    try:
        # Classify with delay between requests
        time.sleep(DELAY_BETWEEN_REQUESTS)
        category = classify_article(article)

        if category in ['Politics', 'Environment', 'Economics']:
            article['news_type'] = category
            progress['filtered'].append(article)
            print(f"âœ… Classified as: {category}")
        else:
            progress['failed'].append(article)
            print("âŒ Not relevant")

    except Exception as e:
        progress['failed'].append(article)
        print(f"âš ï¸ Critical error: {str(e)}")

    # Update and save progress
    progress['last_processed'] = i
    if (i + 1) % SAVE_INTERVAL == 0 or (i + 1) == total_articles:
        with open('filtered_news.json', 'w') as f:
            json.dump(progress['filtered'], f, indent=2)
        with open('failed_articles.json', 'w') as f:
            json.dump(progress['failed'], f, indent=2)
        with open('progress.json', 'w') as f:
            json.dump({'last_processed': i, 'filtered_count': len(progress['filtered'])}, f)
        print(f"ðŸ’¾ Saved progress (processed {i + 1}/{total_articles})")

print("\nProcessing complete!")
print(f"Successfully filtered: {len(progress['filtered'])} articles")
print(f"Failed/discarded: {len(progress['failed'])} articles")
```
