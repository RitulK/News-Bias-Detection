```python
from sentence_transformers import SentenceTransformer
import torch
```

```python
from datasets import load_dataset
dataset = load_dataset("PhilipMay/stsb_multi_mt", "de")
train_dataset = dataset["train"]
eval_dataset = dataset ["dev"]
test_dataset = dataset["test"]

from datasets import Dataset

# Convert examples into HuggingFace-compatible format
train_data = [{"text": item["sentence1"], "text_pair": item["sentence2"]} for item in dataset["train"]]
eval_data = [{"text": item["sentence1"], "text_pair": item["sentence2"]} for item in dataset["dev"]]

train_dataset = Dataset.from_list(train_data)
eval_dataset = Dataset.from_list(eval_data)

```

```python
model = SentenceTransformer('ritulk/MPNET-fine-tuned-political-clustering')
```

```python
from sentence_transformers import SentenceTransformerTrainer, SentenceTransformerTrainingArguments
from sentence_transformers. losses import MultipleNegativesRankingLoss
from sentence_transformers.training_args import BatchSamplers
```

```python
# 3. Define a loss function
loss = MultipleNegativesRankingLoss(model)
```

# with change in loss function

```python
# ✅ Include similarity_score and normalize it to 0–1
train_data = [
    {
        "text": item["sentence1"],
        "text_pair": item["sentence2"],
        "score": item["similarity_score"] / 5.0  # Normalize from 0–5 to 0–1
    }
    for item in dataset["train"]
]

eval_data = [
    {
        "text": item["sentence1"],
        "text_pair": item["sentence2"],
        "score": item["similarity_score"] / 5.0
    }
    for item in dataset["dev"]
]
from datasets import Dataset
train_dataset = Dataset.from_list(train_data)
eval_dataset = Dataset.from_list(eval_data)
```

```python
train_dataset
```

```python
train_dataset.to_pandas()
```

```python
from sentence_transformers.losses import CosineSimilarityLoss
```

```python
loss = CosineSimilarityLoss(model=model)
```

```python
args = SentenceTransformerTrainingArguments(
  # Required parameter:
  output_dir = "models/mpnet-base-all-nli-triplet",
  # Optional*training parameters:
  num_train_epochs=5,
  per_device_train_batch_size=16,
  per_device_eval_batch_size=16,
  warmup_ratio=0.1,
  fp16=True, # Set to False if you get an error that your GPU can't run on FP16
  bf16=False, # Set to True if you have a GPU that supports BF16
  batch_sampler=BatchSamplers.NO_DUPLICATES, # MultipleNegativesRankingLoss benefits from no duplicate samples in a batch
  # Optional tracking/debugging parameters:
  eval_strategy="steps",
  eval_steps=100,
  save_strategy="steps",
  save_steps=100,
  save_total_limit=2,
  logging_steps=100,
  run_name="mpnet-finetuned_rk", # Will be used in W&B if `wandb' is installed
)
```

```python
# Normalize similarity scores from 0–5 to 0–1
def normalize_score(example):
    example["score"] = example["score"] / 5.0
    return example

train_dataset = train_dataset.map(normalize_score)
eval_dataset = eval_dataset.map(normalize_score)
```

```python
trainer = SentenceTransformerTrainer(
  model=model,
  args=args,
  train_dataset=train_dataset,
  eval_dataset=eval_dataset,
  loss=loss
)
```

```python
trainer.train()
```

```python
from sentence_transformers.evaluation import EmbeddingSimilarityEvaluator

# Step 1: Extract sentence pairs and scores
sentences1 = [example["sentence1"] for example in dataset["dev"]]
sentences2 = [example["sentence2"] for example in dataset["dev"]]
scores = [example["similarity_score"] for example in dataset["dev"]]

# Step 2: Initialize the evaluator
evaluator = EmbeddingSimilarityEvaluator(sentences1, sentences2, scores)

# Step 3: Run the evaluation
model.evaluate(evaluator)
```
